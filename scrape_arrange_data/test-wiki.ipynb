```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
# Initialize the Selenium WebDriver (in this case, using Chrome)
driver = webdriver.Chrome()
url = "https://te.wiktionary.org/w/index.php?title=వర్గం:తెలుగు_పదాలు&from=అ"
driver.get(url)

# Locate the div with the ID "mw-pages" that contains the list of words
word_div = driver.find_element(By.ID, "mw-pages")

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(word_div.get_attribute("outerHTML"), "html.parser")

# Extract words from the ul and li elements
word_list = []
for ul in soup.find_all("ul"):
    for li in ul.find_all("li"):
        word = li.get_text()
        word_list.append(word)

print(len(word))
# # Print the list of words
# for word in word_list:
#     print(word)

# Find the anchor tag containing the link to the "తరువాతి పేజీ"
next_page_link = soup.find("a", title="వర్గం:తెలుగు పదాలు", text="తరువాతి పేజీ")
# title="వర్గం:తెలుగు పదాలు">తరువాతి పేజీ
# Get the href attribute of the link
if next_page_link:
    # Get the URL of the next page
    next_page_url = next_page_link.get("href")

    # Construct the full URL
    full_next_page_url = f"https://te.wiktionary.org{next_page_url}"

    # Print the URL of the next page
    print("Link to Next Page:", full_next_page_url)

    # Navigate to the next page
    driver.get(full_next_page_url)

    # Wait for the page to load (you may need to adjust the sleep time)
    time.sleep(4)

    # Extract the text from the next page (modify as needed)
    next_page_text = driver.find_element(By.CSS_SELECTOR, "div.mw-content-ltr").text
    print("Text from Next Page:", next_page_text)
else:
    print("Next Page link not found.")


# Close the WebDriver
# driver.quit()
```
